```python
import os
import shutil
import random

# Paths
images_dir = "/content/coco_yolo/images"
labels_dir = "/content/coco_yolo/labels"

# Create train and val directories
train_images_dir = "/content/coco_yolo/train/images"
val_images_dir = "/content/coco_yolo/val/images"
train_labels_dir = "/content/coco_yolo/train/labels"
val_labels_dir = "/content/coco_yolo/val/labels"

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# List all images
all_images = [f for f in os.listdir(images_dir) if f.endswith(".jpg")]

# Shuffle and split
random.shuffle(all_images)
split = int(0.8 * len(all_images))
train_images = all_images[:split]
val_images = all_images[split:]

# Move images and corresponding labels
for img in train_images:
    shutil.move(os.path.join(images_dir, img), train_images_dir)
    lbl = img.replace(".jpg", ".txt")
    if os.path.exists(os.path.join(labels_dir, lbl)):
        shutil.move(os.path.join(labels_dir, lbl), train_labels_dir)

for img in val_images:
    shutil.move(os.path.join(images_dir, img), val_images_dir)
    lbl = img.replace(".jpg", ".txt")
    if os.path.exists(os.path.join(labels_dir, lbl)):
        shutil.move(os.path.join(labels_dir, lbl), val_labels_dir)

print("Train/Val split done!")
```

```python
import os
import shutil
import random

# Original images/labels
images_val_dir = "/content/coco_yolo/images/val"
labels_val_dir = "/content/coco_yolo/labels/val"

# Train/val directories YOLO expects
train_images_dir = "/content/coco_yolo/train/images"
train_labels_dir = "/content/coco_yolo/train/labels"
val_images_dir = "/content/coco_yolo/val/images"
val_labels_dir = "/content/coco_yolo/val/labels"

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# List all images
all_images = [f for f in os.listdir(images_val_dir) if f.endswith(".jpg")]
random.shuffle(all_images)

# 80% train, 20% val
split_idx = int(0.8 * len(all_images))
train_images = all_images[:split_idx]
val_images = all_images[split_idx:]

# Move train images & labels
for img in train_images:
    shutil.move(os.path.join(images_val_dir, img), os.path.join(train_images_dir, img))
    label_file = img.replace(".jpg", ".txt")
    if os.path.exists(os.path.join(labels_val_dir, label_file)):
        shutil.move(os.path.join(labels_val_dir, label_file), os.path.join(train_labels_dir, label_file))

# Move val images & labels
for img in val_images:
    shutil.move(os.path.join(images_val_dir, img), os.path.join(val_images_dir, img))
    label_file = img.replace(".jpg", ".txt")
    if os.path.exists(os.path.join(labels_val_dir, label_file)):
        shutil.move(os.path.join(labels_val_dir, label_file), os.path.join(val_labels_dir, label_file))

print("Train/Val folders are now populated correctly!")
```

```python
import os, random, shutil, glob

files = glob.glob("/content/*.jpg")

random.shuffle(files)

split = int(len(files) * 0.8)

train_files = files[:split]
val_files = files[split:]

for f in train_files:
    name = os.path.basename(f)
    txt = name.replace(".jpg", ".txt")

    shutil.copy(f, "/content/bdd_yolo/images/train/" + name)
    shutil.copy("/content/" + txt, "/content/bdd_yolo/labels/train/" + txt)

for f in val_files:
    name = os.path.basename(f)
    txt = name.replace(".jpg", ".txt")

    shutil.copy(f, "/content/bdd_yolo/images/val/" + name)
    shutil.copy("/content/" + txt, "/content/bdd_yolo/labels/val/" + txt)

print("Done")
```

```python
import glob
import shutil
import os

src = "/content/Data/train"
dst_img = "/content/bdd_yolo/images/train"
dst_lbl = "/content/bdd_yolo/labels/train"

os.makedirs(dst_img, exist_ok=True)
os.makedirs(dst_lbl, exist_ok=True)

imgs = glob.glob(src + "/*.jpg")

print("Found images:", len(imgs))

for img in imgs:

    name = os.path.basename(img)
    txt = name.replace(".jpg", ".txt")

    shutil.copy(img, dst_img + "/" + name)

    label_path = src + "/" + txt

    if os.path.exists(label_path):
        shutil.copy(label_path, dst_lbl + "/" + txt)

print("Done copying")
```
